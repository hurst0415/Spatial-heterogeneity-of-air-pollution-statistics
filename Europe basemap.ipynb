```python
# libraries
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,mark_inset)
from matplotlib import gridspec
from matplotlib.patches import ConnectionPatch
```

```python
#read dataframe
no_qvalue_df=pd.read_csv('no q,lamda dataframe.csv')
no2_qvalue_df=pd.read_csv('no2 q,lamda dataframe.csv')
pm25_qvalue_df=pd.read_csv('pm2.5 q,lamda dataframe.csv')
pm10_qvalue_df=pd.read_csv('pm10 q,lamda dataframe.csv')
```

```python
#read data_meta
europe_meta=pd.read_csv('data_sites.csv')
#add longitude/latitude to dataframe
no_map_df = no_qvalue_df   
no_map_df['longitude']=europe_meta['longitude']
no_map_df['latitude']=europe_meta['latitude']

no2_map_df = no2_qvalue_df   
no2_map_df['longitude']=europe_meta['longitude']
no2_map_df['latitude']=europe_meta['latitude']

pm25_map_df = pm25_qvalue_df   
pm25_map_df['longitude']=europe_meta['longitude']
pm25_map_df['latitude']=europe_meta['latitude']

pm10_map_df = pm10_qvalue_df   
pm10_map_df['longitude']=europe_meta['longitude']
pm10_map_df['latitude']=europe_meta['latitude']
```

```python
#All sites with valid q,l
frames = [no_map_df, no2_map_df, pm25_map_df, pm10_map_df] 
map_df = pd.concat(frames)
map_df = map_df[map_df["q"]>0]
map_df = map_df.drop('q', 1)
map_df = map_df.drop('l', 1)
map_df=map_df.drop_duplicates()
```

```python
# importing required libraries
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np

# create a figure
fig = plt.figure()

# to change size of subplot's
fig.set_figheight(26)
fig.set_figwidth(34)

# create grid for different subplots
spec = gridspec.GridSpec(ncols=2, nrows=4,
						width_ratios=[1, 1], wspace=0.1,
						hspace=0.7)

#Import location of one site
importdata1 = pd.read_csv('at60178.csv')

# ax0 will take 0th position 
ax0 = fig.add_subplot(spec[:-2, :])
map=Basemap(llcrnrlon=-33, llcrnrlat=25,urcrnrlon=47,urcrnrlat=70.55,ax=ax0)
map.drawmapboundary(fill_color='#A6CAE0')
map.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
map.drawcoastlines(linewidth=0.2, zorder = 0)

# Add a marker per city of the data frame!
x_offset=3
y_offset=1.5
lon = map_df['longitude']
lat = map_df['latitude']
# convert to map projection coords.
for i in map_df.index:
    xpt,ypt = map(lon[i],lat[i])
    # convert back to lat/lon
    lonpt, latpt = map(xpt,ypt,inverse=True)
    map.plot(xpt,ypt,'bo',markersize=8)  # plot blue dots there
xpt1,ypt1 = map(15.628392,47.215703)
# convert back to lat/lon
lonpt1, latpt1 = map(xpt1,ypt1,inverse=True)
map.plot(xpt1,ypt1,'o',markersize=20,color='red')

# ax1 will take "spec[1]"
ax1 = fig.add_subplot(spec[2,0])
oneDay=24
#extract only the day and month
data1 = importdata1[importdata1["variable"]=="no"]["value"]
shortenedDateTime = importdata1[importdata1["variable"]=="no"]["date"]
data1=data1[data1>0]
data_no=np.array(data1)

data2 = importdata1[importdata1["variable"]=="no2"]["value"]
shortenedDateTime = importdata1[importdata1["variable"]=="no2"]["date"]
data2=data2[data2>0]
data_no2=np.array(data2)

ax1.plot(data_no,color='teal',alpha=1,label='$NO$')
ax1.plot(data_no2,color='gold',alpha=0.7,label='$NO_2$')

plt.xticks(np.arange(0,len(shortenedDateTime),183*oneDay),shortenedDateTime[0::183*oneDay])
plt.xticks(fontsize=25,rotation=30)
plt.yticks(fontsize=27)
ax1.tick_params(axis='y', length=6, width=2, colors='k')
ax1.tick_params(axis='x', length=8, width=2, colors='k')
ax1.set_ylim(0,400)
ax1=plt.gca()
ax1.set_xticklabels(item.get_text()[:10] for item in ax1.get_xticklabels())

# ax2 will take "spec[2]"
ax2 = fig.add_subplot(spec[2,1])
#plot PDF 
#sns.distplot(data_no, kde_kws={"color": "green"}, hist=True, ax=ax2)
sns.kdeplot(data_no, color= "teal", bw_adjust=2, clip=(0.2,400),linewidth=2,gridsize=10000,log_scale=(False,True),ax=ax2)
#sns.distplot(data_no2, kde_kws={"color": "gold"}, hist=True, ax=ax2)  
sns.kdeplot(data_no2, color= "gold", bw_adjust=2,  clip=(0.2,105),linewidth=2.3, gridsize=10000,log_scale=(False,True),ax=ax2 )
ax2.set_ylabel('')   
ax2.set_yscale('log')
#ax2.set_xlim(left=0)
ax2.set_ylim(top=10, bottom=10**(-6))
ax2.yaxis.tick_right()
ax2.tick_params(axis='y', length=6, width=2, colors='k')
ax2.tick_params(axis='x', length=8, width=2, colors='k')
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)


# ax3 will take "spec[3]"
ax3 = fig.add_subplot(spec[3,0])

#extract only the day and month
data3 = importdata1[importdata1["variable"]=="pm2.5"]["value"]
shortenedDateTime = importdata1[importdata1["variable"]=="pm2.5"]["date"]
data3=data3[data3>0]
data_pm25=np.array(data3)
data4 = importdata1[importdata1["variable"]=="pm10"]["value"]
shortenedDateTime = importdata1[importdata1["variable"]=="pm10"]["date"]
data4=data4[data4>0]
data_pm10=np.array(data4)

ax3.plot(data_pm10,color='blue',alpha=1,label='$PM_{10}$')
ax3.plot(data_pm25,color='salmon',alpha=0.8,label='$PM2.5')
plt.xticks(np.arange(0,len(shortenedDateTime),183*oneDay),shortenedDateTime[0::183*oneDay])
plt.xticks(fontsize=25,rotation=30)
plt.yticks(fontsize=27)
ax3.tick_params(axis='y', length=6, width=2, colors='k')
ax3.tick_params(axis='x', length=8, width=2, colors='k')
ax3.set_ylim(0,435)
ax3=plt.gca()
ax3.set_xticklabels(item.get_text()[:10] for item in ax3.get_xticklabels())

ax4 = fig.add_subplot(spec[3,1])

#plot PDF 
#sns.distplot(data_pm25, kde={"color": "salmon", "bw":1.5}, hist=True, ax=ax4)
sns.kdeplot(data_pm25, color= "salmon", bw_adjust=2, linewidth=2, clip=(0.2,360),gridsize=100,log_scale=(False,True),ax=ax4 )
#sns.distplot(data_pm10, kde={"color": "blue", "bw":1.5}, hist=True, ax=ax4) 
sns.kdeplot(data_pm10, color= "blue", bw_adjust=2, linewidth=2, clip=(0.2,450),gridsize=100,log_scale=(False,True),ax=ax4 )
ax4.set_ylabel('')   
ax4.set_yscale('log')
#ax4.set_xlim(left=0)
ax4.set_ylim(top=1, bottom=10**(-6))
ax4.yaxis.tick_right()
ax4.tick_params(axis='y', length=6, width=2, colors='k')
ax4.tick_params(axis='x', length=8, width=2, colors='k')
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)

#labels
fig.text(0.3, 0.04, 'Date', ha='center',fontsize="30")
fig.text(0.7, 0.04, 'Concentration [$\mu g m^{-3}$]', ha='center',fontsize="30")
fig.text(0.08, 0.3, 'Concentration [$\mu g m^{-3}$]', va='center', rotation='vertical',fontsize='30')
fig.text(0.93, 0.3, 'Density', va='center', rotation='vertical',fontsize='30')

#legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles[0:], labels=labels[0:],loc='upper center', bbox_to_anchor=(1, 1.3),ncol=3, fancybox=True, shadow=True,fontsize='29')
handles, labels = ax3.get_legend_handles_labels()
ax3.legend(handles=handles[0:], labels=labels[0:],loc='upper center', bbox_to_anchor=(1, 1.3),ncol=3, fancybox=True, shadow=True,fontsize='29')

#plot arrows
con = ConnectionPatch(
    # in axes coordinates
    xyA =(0.595, 0.485), coordsA = ax0.transAxes,
    # x in axes coordinates, y in data coordinates
    xyB =(0.5, 430), coordsB = ax1.get_yaxis_transform(),
    arrowstyle ="->")
ax3.add_artist(con)
con.set_linewidth(3)

con = ConnectionPatch(
    # in axes coordinates
    xyA =(0.62, 0.485), coordsA = ax0.transAxes,
    # x in axes coordinates, y in data coordinates
    xyB =(0.5, 57), coordsB = ax2.get_yaxis_transform(),
    arrowstyle ="->")
ax3.add_artist(con)
con.set_linewidth(3)

ax0.text(0.95, 0.9,('a'), transform=ax0.transAxes, size=35)
ax1.text(0.95, 0.85,('b'), transform=ax1.transAxes, size=35)
ax2.text(0.95, 0.85,('c'), transform=ax2.transAxes, size=35)
ax3.text(0.95, 0.85,('d'), transform=ax3.transAxes, size=35)
ax4.text(0.95, 0.85,('e'), transform=ax4.transAxes, size=35)

plt.savefig('map fig.pdf',dpi=75,bbox_inches='tight', pad_inches = 0.05)
plt.show()
```
